```python
import numpy as np


def task_6():
    arr = np.array([1, 2, 3, 4, 5])
    new_arr = np.where(arr != 0, np.array([-1]), arr)
    print('Array: ', new_arr)

    arr = np.array([0, 0, 0, 0, 0])
    new_arr = np.where(arr != 0, np.array([-1]), arr)
    print('Array: ', new_arr)

    arr = np.array([0, 1, 2, 3, 0, 4, 0, 0, 5, 6])
    new_arr = np.where(arr != 0, np.array([-1]), arr)
    print('Array: ', new_arr)


def task_7():
    arr = np.array([1, 2, 3, 4, 5])
    new_arr = np.where(arr == 0, np.array([-1]), arr)
    print('Array: ', new_arr)

    arr = np.array([0, 0, 0, 0, 0])
    new_arr = np.where(arr == 0, np.array([-1]), arr)
    print('Array: ', new_arr)

    arr = np.array([0, 1, 2, 3, 0, 4, 0, 0, 5, 6])
    new_arr = np.where(arr == 0, np.array([-1]), arr)
    print('Array: ', new_arr)


def task_9():
    n = 5
    arr = np.arange(n, -1, -1)
    print('Array: ', arr)

    n = 3
    arr = np.arange(n, -1, -1)
    print('Array: ', arr)

    n = 10
    arr = np.arange(n, -1, -1)
    print('Array: ', arr)


def task_10():
    m, n = 3, 3
    arr = np.random.rand(m, n)
    print('Array:\n', arr)
    print('min: ', np.round(np.min(arr), 3))
    print('max: ', np.round(np.max(arr), 3))
    print('mean: ', np.round(np.mean(arr), 3))
    print('deviation: ', np.round(np.std(arr), 3))

    m, n = 6, 2
    arr = np.random.rand(m, n)
    print('Array:\n', arr)
    print('min: ', np.round(np.min(arr), 3))
    print('max: ', np.round(np.max(arr), 3))
    print('mean: ', np.round(np.mean(arr), 3))
    print('deviation: ', np.round(np.std(arr), 3))


def task_11():
    n = 3
    x = np.ones((n, n))
    x = np.pad(x, pad_width=1, mode='constant', constant_values=0)
    print(x)

    n = 5
    x = np.ones((n, n))
    x = np.pad(x, pad_width=1, mode='constant', constant_values=0)
    print(x)

    n = 8
    x = np.ones((n, n))
    x = np.pad(x, pad_width=1, mode='constant', constant_values=0)
    print(x)


def task_15():
    # n is twice smaller then board size
    n = 4
    checkerboard = np.tile(np.array([[0, 1], [1, 0]]), (n, n))
    print(checkerboard)

    n = 3
    checkerboard = np.tile(np.array([[0, 1], [1, 0]]), (n, n))
    print(checkerboard)

    n = 5
    checkerboard = np.tile(np.array([[0, 1], [1, 0]]), (n, n))
    print(checkerboard)


def task_17():
    n = 10
    arr = np.arange(n + 1)
    print('Array: ', arr)
    new_arr = np.where(((arr >= n / 2) & (arr <= 3 * n / 4)), np.array([0]), arr)
    print('Replace all values greater than n/2 and less than 3n/4 with zeros: n =', n)
    print(new_arr)

    n = 15
    arr = np.arange(n + 1)
    print('Array: ', arr)
    new_arr = np.where(((arr >= n / 2) & (arr <= 3 * n / 4)), np.array([0]), arr)
    print('Replace all values greater than n/2 and less than 3n/4 with zeros: n =', n)
    print(new_arr)

    n = 5
    arr = np.arange(n + 1)
    print('Array: ', arr)
    new_arr = np.where(((arr >= n / 2) & (arr <= 3 * n / 4)), np.array([0]), arr)
    print('Replace all values greater than n/2 and less than 3n/4 with zeros: n =', n)
    print(new_arr)


def task_18():
    n = 10
    arr = np.arange(n + 1)
    print('Array: ', arr)
    new_arr = np.where(((arr < n / 2) | (arr > 3 * n / 4)), -arr, arr)
    print('Replace all values greater than n/2 and less than 3n/4 with zeros: n =', n)
    print(new_arr)

    n = 15
    arr = np.arange(n + 1)
    print('Array: ', arr)
    new_arr = np.where(((arr < n / 2) | (arr > 3 * n / 4)), -arr, arr)
    print('Replace all values greater than n/2 and less than 3n/4 with zeros: n =', n)
    print(new_arr)

    n = 5
    arr = np.arange(n + 1)
    print('Array: ', arr)
    new_arr = np.where(((arr < n / 2) | (arr > 3 * n / 4)), -arr, arr)
    print('Replace all values greater than n/2 and less than 3n/4 with zeros: n =', n)
    print(new_arr)


def task_20():
    n = 5
    arr = np.round(np.linspace(0.0, 1.0, num=n + 1, endpoint=False)[1:], 3)
    print('Array: ', arr)

    n = 10
    arr = np.round(np.linspace(0.0, 1.0, num=n + 1, endpoint=False)[1:], 3)
    print('Array: ', arr)

    n = 15
    arr = np.round(np.linspace(0.0, 1.0, num=n + 1, endpoint=False)[1:], 3)
    print('Array: ', arr)


def task_22():
    n, m = 4, 8
    arr = np.random.randint(0, m, (n, n))
    print('Initial vector\n', arr)
    new_arr = np.where(arr == np.max(arr), -arr, arr)
    print('Modified vector\n', new_arr)

    n, m = 3, 5
    arr = np.random.randint(0, m, (n, n))
    print('Initial vector\n', arr)
    new_arr = np.where(arr == np.max(arr), -arr, arr)
    print('Modified vector\n', new_arr)

    n, m = 4, 3
    arr = np.random.randint(0, m, (n, n))
    print('Initial vector\n', arr)
    new_arr = np.where(arr == np.max(arr), -arr, arr)
    print('Modified vector\n', new_arr)


print('task 6')
task_6()
print('task 7')
task_7()
print('task 9')
task_9()
print('task 10')
task_10()
print('task 11')
task_11()
print('task 15')
task_15()
print('task 17')
task_17()
print('task 18')
task_18()
print('task 20')
task_20()
print('task 22')
task_22()
```
